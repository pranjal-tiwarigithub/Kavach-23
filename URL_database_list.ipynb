```python
import sqlite3
import pandas as pd
  
# Connect to SQLite database
conn = sqlite3.connect(r'url_data.db')
  
# Load CSV data into Pandas DataFrame
url_info = pd.read_csv('https://raw.githubusercontent.com/pranjal-tiwarigithub/Kavach-23/main/urlhaus.abuse.ch.csv')
```

```python
url_info.head()
```

```python
# Write the data to a sqlite table
url_info.to_sql('url_data', conn, if_exists='replace', index=False)
  
```

```python
# Create a cursor object
cur = conn.cursor()

malicious_url_list = []
# Fetch and display result
for row in cur.execute('SELECT url FROM url_data'):
    malicious_url_list.append(row)
# Close connection to SQLite database
```

```python
malicious_url_list
```

```python
check = input("Enter the link you want to check ")
```

```python
check
```

```python
safety = True 
for i in malicious_url_list:
  if check == i[0]:
    safety = False 
```
